```python
#Pasamos todas las fechas a año
import re

patron = r'(\d{4})'  # Este patrón busca 4 dígitos consecutivos en una cadena cruda (r: raw).

def año(cadena:str) -> str|None:
    """
    La función año toma la cadena de texto y devuelve un año en formato entero
    o None si la cadena está vacía.
    """
    match = re.search(patron, str(cadena))
    if match:
        return match.group(1)
    else:
        return None

dfgames['release_date'] = dfgames['release_date'].apply(año)
```

```python
#importamos la herramienta para hacer el analisis de sentimiento
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

#Descarga de recursos necesarios para VADER
nltk.download('vader_lexicon')

#Creación del objeto SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
dfsent = pd.DataFrame(lista, columns=['item_id', 'review']) #creamos un dataframe nuevo donde vamos a hacer el analisis de sentimiento
```

```python
#le añadimos el puntaje de sentimiento a cada review
dfsent['sentiment'] = dfsent['review'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
```

```python
#Agregamos columnas con 1 y 0 por cada sentimiento para luego poder hacer el conteo
dfsent['sent_neg'] = dfsent['sentiment'].apply(lambda x: 1 if x < 0 else 0)
dfsent['sent_neu'] = dfsent['sentiment'].apply(lambda x: 1 if 0 <= x <= 0.3 else 0)
dfsent['sent_pos'] = dfsent['sentiment'].apply(lambda x: 1 if x > 0.3 else 0)
```

```python
#Agrupamos por id y nos queda un conteo de la cantidad de reviews que tiene cada juego
dfsent.drop(columns=['review', 'sentiment'], inplace=True)
dfsent = dfsent.groupby('item_id').sum()
```

```python
#Aca le agregamos la columna de sentimiento en funcion de la cantidad de review de cada tipo
def colsent(df):
    if df['sent_pos'] > df["sent_neu"] and df['sent_pos'] > df["sent_neg"]:
        return 2
    elif df['sent_neu'] >= df["sent_pos"] and df['sent_neu'] >= df["sent_neg"]:
        return 1
    else:
        return 0
    
dfsent['sentiment'] = dfsent.apply(colsent, axis=1)
```

```python
#Reseteamos el indice y pasamos item_id a int
dfsent = dfsent.reset_index()
dfsent['item_id'] = dfsent['item_id'].astype('int')
```

```python
dfsent
```

```python
#Hacemos una lista con los recommend de cada usuario
lista = []
for i in range(0, len(dfreviews)):
    l = []
    l.append(dfreviews['user_id'][i])
    for elem in dfreviews['reviews'][i]:
        l.append(elem['recommend'])
    lista.append(l)
```

```python
# Extraemos la cantidad de reviews positivas y negativas por usuario
pos = []
neg = []

for i in range(len(lista)):
    x = 0
    z = 0
    for elem in lista[i]:
        if elem == True:
            x = x + 1
        if elem == False:
            z = z + 1
    pos.append(x)
    neg.append(z)
```

```python
dfreviews['recommend_pos'] = pos
dfreviews['recommend_neg'] = neg 
```

```python
#Aca le agregamos la columna con el porcentaje de recomendacion de cada usuario
def percent(df):
    if df['recommend_pos'] == 0 and df['recommend_neg'] == 0:
        return 0
    else:
        return round(df['recommend_pos'] / (df['recommend_pos'] + df['recommend_neg']) * 100, 2)
    
    
dfreviews["% recommend"] = dfreviews.apply(percent, axis=1)
```

```python
#Una vez extraido todo lo necesario, borramos la columna de reviews
dfreviews.drop(columns='reviews', inplace=True)
```
